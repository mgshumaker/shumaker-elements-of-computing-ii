### __Quick Introduction to the Iris Dataset:__

The Iris dataset, often called the "Hello World" of machine learning, is one of the most famous and enduring datasets in data science. Compiled in the 1930s, it remains a standard benchmark for classification, clustering, and visualization tasks.

__FUN FACT:__ The Iris dataset is older than modern computing. The data was collected by American botanist Edgar Anderson in 1935.

____________________________________________________________________________________________________________

Dataset Structure:
* The dataset consists of 150 samples, with 50 samples from three species. 

Four Features (Measurements in cm):
* Sepal Length: The length of the outer green leaf-like structures.
* Sepal Width: The width of the sepals.
* Petal Length: The length of the inner colored petals.
* Petal Width: The width of the petals.

Three Target Species:
* Iris setosa (often encoded as 0).
* Iris versicolor (often encoded as 1).
* Iris virginica (often encoded as 2).

```python
# Group A: Create the ridgeline plot

sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})
iris = sns.load_dataset("iris")

# Initialize the FacetGrid object
g = sns.FacetGrid(iris, row="species", hue="species", aspect=15, height=.5)

# Draw the densities
g.map(sns.kdeplot, "sepal_length", bw_adjust=.5, clip_on=False, fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, "sepal_length", clip_on=False, color="w", lw=2, bw_adjust=.5)

# Add labeling function
def label(x, color, label):
    ax = plt.gca()
    # x=0 is left, y=.2 is slightly above baseline
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)

g.map(label, "sepal_length")

# Final styling to make them overlap
g.figure.subplots_adjust(hspace=-.25)
g.set_titles("")
g.figsize=(10, 6)
g.set(yticks=[], ylabel="")
g.despine(bottom=True, left=True)
plt.show()
```

__Polar Bar Chart__

The Anatomy of the Plot:

* The Circular Axis (Angles): The 360 degrees of the circle are divided into three equal segments, one for each species: Setosa, Versicolor, and Virginica.

* The Radial Axis (Distance from Center): The "height" of each bar is measured by its distance from the center point of the circle. A longer bar indicates a higher mean value.

The Bars:
- Length: Represents the specific value of the mean petal length (in cm).

- Width: The angular width is set to be quite wide to fill the circular space for better visual balance.

```python
# Create a Polar Bar Chart

# Load dataset and calculate means
df_mean = df.groupby("species")["petal_length"].mean()

# Set up the polar coordinates
labels = df_mean.index
values = df_mean.values
num_vars = len(labels)

# Compute angles for each bar
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
# "Close" the circle by appending the first value to the end
values = np.concatenate((values, [values[0]]))
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw the bars
ax.bar(angles[:-1], df_mean.values, color=['#440154', '#21918c', '#fde725'], alpha=0.7, width=1.5)

# Fix labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_title("Mean Petal Length by Species", va='bottom')

plt.show()
```

__Understanding the Results:__

* __Setosa (Shortest Bar):__ The bar is very close to the center. This reflects that Setosa has the smallest average petal length (typically around 1.46cm).
* __Versicolor (Medium Bar):__ The bar extends significantly further than Setosa, showing an average length of roughly 4.26cm.
* __Virginica (Longest Bar):__ This bar reaches the furthest toward the outer edge of the circle, representing the largest average petal length of roughly 5.55cm.

# Section 4: Part of a Whole

```python
# FIRST - conda install plotly

import plotly.express as px

# Load the iris dataset
df = px.data.iris()

# Create a treemap sized by petal_width
# 'path' creates the hierarchy: Species -> Individual Sample ID
fig = px.treemap(df, 
                 path=['species'], 
                 values='petal_width',
                 color='species',
                 title="Iris Treemap: Rectangle Size by Petal Width")

fig.show()
```

#### __What this Treemap Shows:__

* __Hierarchy:__ The largest rectangles represent the three species (Setosa, Versicolor, Virginica).

* __Size (Area):__ The size of each species block reflects the sum of the value you chose (e.g., petal length). Since * * Virginica has the longest petals, its section will be the largest.

* __Individual Samples:__ Inside each species block, you can see smaller rectangles representing individual flower samples.

________________________________________________________________________________________________

A __Sunburst chart__ is essentially a radial treemap (or pie chart). It uses concentric circles to show hierarchy, with the inner circle representing the root and outer rings showing further detail.

In the Iris dataset, a Sunburst chart is often used to visualize the hierarchy of Species → Individual Samples, with the size of each segment determined by a feature like petal width. 

```python
# Create the Sunburst chart
fig = px.sunburst(df, 
                  path=['species'], 
                  values='petal_width',
                  color='species',
                  title="Iris Sunburst: Sized by Petal Width")

fig.show()
```

#### How to Read This Chart
* The Center (Inner Ring): Represents the three Iris species (Setosa, Versicolor, Virginica).

* The Outer Ring: Shows individual flower samples within those species.

* Segment Size: The "arc" or angle of each segment is proportional to its petal width. You will see the Virginica section occupies the largest portion of the circle because it has the widest petals.

##### Referring to the creation of  venn diagram below

This approach calculates the range (min to max) of petal widths for each species and creates a Venn diagram showing how many flowers from each species fall into shared measurement "bins."

```python
# FIRST - pip install matplotlib-venn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Load the dataset
iris = sns.load_dataset('iris')

# Define sets based on Petal Width thresholds or specific values
# Example: Let's see which species share common petal width values
setosa_widths = set(iris[iris['species'] == 'setosa']['petal_width'])
versicolor_widths = set(iris[iris['species'] == 'versicolor']['petal_width'])
virginica_widths = set(iris[iris['species'] == 'virginica']['petal_width'])

# Create the Venn Diagram
plt.figure(figsize=(8, 8))
venn3([setosa_widths, versicolor_widths, virginica_widths], 
      set_labels=('Setosa', 'Versicolor', 'Virginica'),
      set_colors=('skyblue', 'lightgreen', 'salmon'), 
      alpha=0.7)

plt.title("Overlap of Unique Petal Width Values (cm) by Species")
plt.show()
```

#### Explanation:

* The Circles: Each circle represents a species and contains all the unique measurement values for petal width found in that group.

* Setosa (Isolated): You will notice the Setosa circle has zero overlap with the other two. This is because Setosa petal widths are very small (0.1–0.6 cm) and don't match the larger measurements of the other species.

* Versicolor vs. Virginica (Overlap): These two circles will typically overlap. This tells you there are specific petal widths (e.g., 1.4 cm or 1.8 cm) that appear in both species, making that specific measurement a "shared" characteristic.

```python
#FIRST - pip install pywaffle


import matplotlib.pyplot as plt
from pywaffle import Waffle
import seaborn as sns

# Load dataset
iris = sns.load_dataset("iris")

# Count occurrences of each species
data = iris['species'].value_counts().to_dict()

# Create the Waffle Chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10, 
    values=data, 
    colors=["#440154", "#21918c", "#fde725"],
    title={'label': 'Species Distribution in Iris Dataset', 'loc': 'left'},
    labels=[f"{k} ({v})" for k, v in data.items()],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.2), 'ncol': 3, 'frameon': False}
)

plt.show()
```

This script calculates the count of each species and displays them as a 10 x 15 grid, where each icon represents one flower.

Since the Iris dataset contains 50 samples for each of the three species, a waffle chart is an excellent way to visualize this perfectly even distribution.
